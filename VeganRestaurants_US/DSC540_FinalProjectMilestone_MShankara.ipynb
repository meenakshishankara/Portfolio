### Project Milestone - 3 - Perform data transformation and/or cleansing steps to your website data

```python
## import the required packages
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
```

### Scrape the data from website

```python
## scraping the data from happycow.net, all vegan/vegetarian and veg-friendly eateries all over the country
## using the Selenium webdriver package, the page is manually opened in the chrome browser 
## and the html data is stored in a soup object
## credits to Mike Zoucha for this code piece as Captcha was blocking the scraping process.

url = 'https://www.happycow.net/north_america/usa/'
driver = webdriver.Chrome() 
driver.get(url)
## WILL OPEN BROWSER AND MAKE YOU MANUALLY DO CAPTCHA PICTURES
time.sleep(5)
soup = BeautifulSoup(driver.page_source, features='html.parser')
driver.quit()
```

```python
## from the html the state and number of eateries data is saved in a dataframe.

links = []
links = soup("a", "block", href=True)
states = []
numbers = []
for link in links:

    if len(link.text.splitlines()) > 4:
        states.append(link.text.splitlines()[2])
        numbers.append(link.text.splitlines()[3])
d = {'State': states, 'Number': numbers}
df = pd.DataFrame(d)
print(df)
```

### Step 1 - Cleaning the Number column - remove the brackets

```python
for i in range(len(df["Number"])):
    df["Number"][i] = df["Number"][i].replace('(','')
    df["Number"][i] = df["Number"][i].replace(')','')
    
print(df)
```

```python
# there are 50 states + Washington DC in the US. Therefore we will clean the DF accordingly
# we will upload a csv with all the names of the states and then compare with our df

states_df = pd.read_csv("states.csv")
states_df
```

### Step 2 - Remove records that are not states

```python
## filter the data in statesvegan data from based on the the states.csv
filter_state = df["State"].isin(states_df["State"])
statevegan_df = df[filter_state]
print(statevegan_df)
```

### Step 3 - Sort the dataframe on State

```python
statevegan_df.sort_values(by=["State"])
states_df.sort_values(by=["State"])
            
```

### Step 4 - Merge the state codes into the statevegan_df dataframe

```python
#Merge data on state
statevegan_df = pd.merge(statevegan_df, states_df, on='State', how='inner')
statevegan_df
```

### step 5 - formatting the Number column from string to Int

```python
statevegan_df['Number'] = pd.to_numeric(statevegan_df['Number'])
statevegan_df.dtypes
```

```python
## merging data from all the files to load into a table Food_Consolidated 

mergeall_sql = pd.read_sql("""select count(*) as Number, categories, state as StateCode
                                from (Select distinct
                                   ff.id,

                                   'Eatery' as categories, ff.province as state
                                   from US_FastFood ff 
                                   
                       
                                   UNION 
                       
                                   Select distinct
                                   r.id,
                                   
                                   'Eatery' as categories, r.state
                                   from US_VegRestaurant r
                                   )
                                   
                                group by categories, state
                                
                                union
                                
                                select Number, 'HappyCow Veg' as categories, StateCode 
                                FROM State_VeganRestaurant""", conn)

mergeall_sql.to_sql('Food_Consolidated',conn)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

## Visualization 1 
## Which state has the most Vegan/Veg/Veg-friendly establishments
## we will plot on the Food_Consolidated table that is the merged data from all the source

Category_Veg = pd.read_sql("""Select sum(Number) Count, StateCode from Food_Consolidated 
                            WHERE categories like '%Veg%'
                            group by StateCode
                            order by sum(Number) desc LIMIT 15""", conn)
Category_Veg
```

```python
Category_Veg.plot.bar(y = 'Count', x='StateCode')
plt.ylabel('vegan/veg/veg-friendly eateries')
plt.title('Vegan/Veg/Veg-friendly Eateries')
plt.show()
```

```python
## Visualization 2
## Which states have the highest number of fast foods
fastfood_rest = pd.read_sql('select count(distinct id) as FastFood, province from US_FastFood GROUP BY Province order by FastFood desc LIMIT 10', conn)
strfastfood = fastfood_rest['FastFood'].apply(str)

plt.pie(fastfood_rest['FastFood'], labels = ( fastfood_rest['province'] + ' : ' + strfastfood), rotatelabels = True)
plt.show()
fastfood_rest
```

```python
## Visualization 3

## Year when restaurants were added in the system 
import numpy as np

dateadded_USAll = pd.read_sql("""select count(*) as restadded, 
                            cast(dateAdded as YEAR) yearadded
                                from (Select distinct
                                   ff.id,
                                   ff.dateAdded
                                   
                                   from US_FastFood ff
                       
                                   UNION 
                       
                                   Select distinct
                                   r.id,
                                   r.dateAdded
                                   from US_VegRestaurant r)
                                group by yearadded order by yearadded desc
                                """,conn)
 
    
dateadded_USAll.plot(x='yearadded',y='restadded' )
plt.show()
dateadded_USAll

##select distinct id, dateAdded, cast(dateAdded as YEAR) as year from US_VegRestaurant) group by year order by year desc
```

```python
## Visualization 4
## different categories/cuisines of restaurants that serve Vegetarian friendly food across the country

category_USAll = pd.read_sql("""select count(*) as number, categories
                                from (Select distinct
                                   ff.id,

                                   ff.categories
                                   from US_FastFood ff
                       
                                   UNION 
                       
                                   Select distinct
                                   r.id,
                                   
                                   r.categories
                                   from US_VegRestaurant r)
                                   where categories like ('%Veg%')
                                group by categories
                                having number > 1""",conn)


category_USAll.plot.barh(y='number',x='categories')
#plt.ylabel('categories')
#plt.xlabel('number')
plt.title('Vegan/Veg/Veg-friendly Eateries')
plt.show()

category_USAll

```
